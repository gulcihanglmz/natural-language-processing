**N-Gram: Bir dil modelinde kullanılan kelime veya karakter dizisinin uzunluğunu belirtir.**

**N-Gram Modelleri: Metinleri n kelimelik veya n karakterlik kısımlara bölerek analiz eder.**

Uni-gram, bi-gram, tri-gram

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = ["This is a ngram notebook.",
             "This is a nature language models notebook."]
```

```python
# unigram, bigram, trigram modelleri
vectorizer_unigram = CountVectorizer(ngram_range=(1,1))
vectorizer_bigram = CountVectorizer(ngram_range=(2,2))
vectorizer_trigram = CountVectorizer(ngram_range=(3,3))
```

```python
# unigram
x_unigram = vectorizer_unigram.fit_transform(documents)
unigram_feature_names = vectorizer_unigram.get_feature_names_out()
print(unigram_feature_names)
```

```python
# bigram
x_bigram= vectorizer_bigram.fit_transform(documents)
bigram_feature_names = vectorizer_bigram.get_feature_names_out()
print(bigram_feature_names)
```

```python
# trigram
x_trigram = vectorizer_trigram.fit_transform(documents)
trigram_feature_names = vectorizer_trigram.get_feature_names_out()
print(trigram_feature_names)
```
